```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import pandas as pd


def train(X):
    # get the nearest neighbors (brute, ball_tree, kd_tree, auto)
    nbrs = NearestNeighbors(n_neighbors=8, algorithm='kd_tree').fit(X)
    # return distances and indices (need explain)
    distances, indices = nbrs.kneighbors(X)
    ar = nbrs.kneighbors_graph(X).toarray()
    return distances, indices, ar

# ball tree algorithm
# 1,2,4,2,6,4,2
#


def test(train, test):
    frames = [train, test]
    result = pd.concat(frames)
    # get the nearest neighbors (brute, ball_tree, kd_tree, auto)
    nbrs = NearestNeighbors(n_neighbors=30, algorithm='brute').fit(result)
    # return distances and indices
    distances, indices = nbrs.kneighbors(result)
    return indices[len(train):len(result)-1]


def label_quality(indices, order):
    # label the quality
    label_list = []
    for i in range(0, len(indices)):
        label = [indices[i][0]]
        for j in range(1, len(indices[i])):
            label.append(order[indices[i][j]])
        label_list.append(label)
    return label_list


def max_count(label_list):
    # find the max frequency
    count_list = []
    for i in range(0, len(label_list)):
        cnt = [label_list[i][0]]
        ls = label_list[i][1:len(label_list)]
        cnt.append(max(ls, key=ls.count))
        count_list.append(cnt)
    return count_list


def check_correction(label, y):
    cor = 0
    incor = 0
    for i in range(0, len(label)):
        real = y[label[i][0]]
        pred = label[i][1]
        if real == pred:
            cor += 1
        else:
            incor += 1

    print(cor)
    print(incor)
    cor_pre = (cor)/(cor+incor)
    incor_pre = (incor)/(cor+incor)
    return cor_pre, incor_pre

```
